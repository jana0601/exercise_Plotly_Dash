```python
from jupyter_dash import JupyterDash
from dash import html, dcc, Input, Output

import plotly.express as px
import pandas as pd

#  data
df = px.data.iris()

# Initialize app
#app = Dash(__name__)
app = JupyterDash(__name__)  # <- Not dash.Dash

# Layout
app.layout = html.Div([
    html.H1("Iris Dataset Scatter Plot"),
    dcc.Dropdown(
        id='species-dropdown',
        options=[{'label': s, 'value': s} for s in df['species'].unique()],
        value='setosa'
    ),
    dcc.Graph(id='scatter-plot')
])

# Callback to update plot
@app.callback(
    Output('scatter-plot', 'figure'),
    Input('species-dropdown', 'value')
)
def update_graph(selected_species):
    filtered = df[df['species'] == selected_species]
    fig = px.scatter(filtered, x='sepal_width', y='sepal_length', color='species')
    return fig

# Run app
if __name__ == '__main__':
    app.run(jupyter_mode="external")
```
